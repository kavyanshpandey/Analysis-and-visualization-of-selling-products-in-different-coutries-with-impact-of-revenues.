```python
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
dataset = pd.read_csv("https://raw.githubusercontent.com/devbabar/sales_data_analysis_visualization/master/WA_Sales_Products_2012-14.csv")
```

```python
dataset.head()
```

```python
dataset.describe()
```

```python
dataset.info()
```

```python
ess_dataset = dataset[['Retailer country','Product type', 'Quantity']]
```

```python
category_group = ess_dataset.groupby(['Retailer country','Product type']).sum()
```

```python
category_group
```

```python
category_group.unstack()
```

```python
category_group_graph = category_group.unstack().plot(kind='bar',stacked= True, legend=('0'),figsize=(14,14))
```
